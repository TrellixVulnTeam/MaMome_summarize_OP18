```python
import os
import csv
import pandas as pd
```

```python
def raw_metrics(doc_name):

    raw_name = doc_name[:-4] + ".txt"

    with open("../docproc/TxtData/" + raw_name) as f:
        raw = f.read()

    raw = raw.lower()
    rawterms = set(raw.split(" "))
    rawcommterms = len(common_member(bacwords, rawterms))

    rawsents = set(raw.split('.'))
    raw_relsents = rel_sents(rawsents, bacwords)

    raw_tup = (rawcommterms, raw_relsents)

    return raw_tup


def sum_metrics(doc_name, suffix, bacwords, bert_type, raw_tup):

    summary_name = doc_name[:-4] + suffix + ".txt"

    if bert_type == "General":
        with open("Summary_BL/" + summary_name, "r") as f:
            summary = f.read()
    else:
        with open("Summary_PM/" + summary_name, "r") as f:
            summary = f.read()

    summary = summary.lower()
    sumterms = set(summary.split(" "))
    sumcomterms = len(common_member(bacwords, sumterms))

    sumsents = set(summary.split())
    sum_relsents = rel_sents(sumsents, bacwords)

    if raw_tup[0] != 0:
        prop_terms = sumcomterms / raw_tup[0]
    else:
        prop_terms = None

    if raw_tup[1] != 0:
        prop_sents = sum_relsents / raw_tup[1]
    else:
        prop_sents = None

    met_list = [suffix, prop_terms, prop_sents]

    return met_list


def common_member(a, b):
    a_set = set(a)
    b_set = set(b)

    commterms = ""

    if (a_set & b_set):
        commterms = list(a_set & b_set)

    return commterms


def rel_sents(sents, bacwords):

    counter = 0

    for sent in sents:
        temp = sent.lower()
        if any(check in temp.split(' ') for check in bacwords):
            counter += 1

    return counter


```

```python
with open('../docproc/bac_list.csv') as f:
    reader = csv.reader(f)
    blist = list(reader)

bacteria = []

for bac in blist[2:]:
    if len(bac) > 0:
         bacteria += bac[0].split(' ')

bacwords = [word.lower() for word in bacteria if (word.isalpha() and len(word) > 1)]
```

```python

articles = sorted(os.listdir("../MaMomePDFs"))

met_dict = {}
```

```python
gen_met_dict = {}

for article in articles:
    raw_tup = raw_metrics(article)
    suffices = ["_targ_verb", "_full_verb"]
    for suffix in suffices:
        art_mets = sum_metrics(article, suffix, bacwords, "General", raw_tup)
        key = article[:-4] + "_" + art_mets[0]
        metlist = ["General"]
        metlist.append(art_mets[0])
        metlist.append(art_mets[1])
        metlist.append(art_mets[2])
        gen_met_dict[key] = metlist
```

```python
gen_df = pd.DataFrame.from_dict(gen_met_dict, orient = "index")

gen_df
```

```python
pm_met_dict = {}

for article in articles:
    raw_tup = raw_metrics(article)
    suffices = ["_targ_verb", "_full_verb"]
    for suffix in suffices:
        art_mets = sum_metrics(article, suffix, bacwords, "PubMed", raw_tup)
        key = article[:-4] + "_" + art_mets[0]
        metlist = ["PubMed"]
        metlist.append(art_mets[0])
        metlist.append(art_mets[1])
        metlist.append(art_mets[2])
        pm_met_dict[key] = metlist
```

```python

pm_df = pd.DataFrame.from_dict(pm_met_dict, orient='index')
pm_df
```

```python
gen_df.columns = ["BERT", "SummaryType", "PropBacTerms", "PropBacSents"]
pm_df.columns = ["BERT", "SummaryType", "PropBacTerms", "PropBacSents"]
```

```python
pm_drop = pm_df[pm_df.PropBacTerms > 1] + pm_df[pm_df.PropBacSents > 1]
gen_drop = gen_df[gen_df.PropBacTerms > 1] + gen_df[gen_df.PropBacSents > 1]
```

```python
todrop = list(set(list(pm_drop.index) + list(gen_drop.index)))
```

```python
pm_df_clean = pm_df.drop(index = todrop)
pm_df_clean = pm_df_clean.dropna(axis=0)

gen_df_clean = gen_df.drop(index = todrop)
gen_df_clean = gen_df_clean.dropna(axis=0)
```

```python
pmstats = pm_df_clean.groupby("SummaryType")[['PropBacTerms','PropBacSents']].mean()
```

```python
genstats = gen_df_clean.groupby("SummaryType")[['PropBacTerms','PropBacSents']].mean()
```

```python
pmstats["BERT_Type"] = ["MSR PubMed BERT","MSR PubMed BERT"]
pmstats["Summary_Type"] = ["Full Verbose","Targeted Verbose"]

genstats["BERT_Type"] = ["Large BERT", "Large BERT"]
genstats["Summary_Type"] = ["Full Verbose","Targeted Verbose"]
```

```python
pmstats = pmstats[["BERT_Type", "Summary_Type", "PropBacTerms", "PropBacSents" ]]
genstats = genstats[["BERT_Type", "Summary_Type", "PropBacTerms", "PropBacSents" ]]
```

```python
pd.concat([pmstats, genstats])
```
